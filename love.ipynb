 Q3. Replace missing values in the Quantity and Price columns with their respective mean values using
Pandas.


-To replace missing values in **Quantity** and **Price** with their respective mean values, you can use `fillna()` along with `mean()` in Pandas.

---

### ✅ Python code (Q3 solution)

```python id="impute_q3"
# Ensure columns are numeric (in case of object dtype due to blanks)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Fill missing values with mean
df["Quantity"].fillna(df["Quantity"].mean(), inplace=True)
df["Price"].fillna(df["Price"].mean(), inplace=True)

# Verify no missing values remain in these columns
print(df[["Quantity", "Price"]].isnull().sum())
```

---

 What this does

* Converts **Quantity** and **Price** into numeric format (important because blanks become strings otherwise)
* Computes mean of each column
* Replaces missing values with:

  * Mean Quantity → for missing Quantity values
  * Mean Price → for missing Price values

---

Expected result

After execution:

```text
Quantity    0
Price       0

 Q4. Remove all rows where the Product Category or Region column contains missing values.


   -To remove all rows where **Product Category** or **Region** has missing values, you can use `dropna()` with the `subset` parameter in Pandas.

---

### ✅ Python code (Q4 solution)

```python id="q4_dropna"
# Drop rows where Product Category or Region is missing
df = df.dropna(subset=["Product Category", "Region"])

# Reset index after dropping rows (optional but recommended)
df.reset_index(drop=True, inplace=True)

# Check remaining missing values
print(df[["Product Category", "Region"]].isnull().sum())
print("\nRemaining dataset shape:", df.shape)
```

---

### 📌 What this does

* `subset=["Product Category", "Region"]` ensures:

  * If either column is missing → row is removed
* `dropna()` cleans those incomplete records
* `reset_index()` keeps DataFrame tidy after deletions

---

* Rows with blank **Product Category** (e.g., T013, T036, T075, etc.)

---

### 📌 What this does

* `groupby()` → calculates total revenue per category
* `sort_values(ascending=False)` → ranks from highest to lowest
* `.head(3)` → selects top 3 categories
* `ascending=True` → gives lowest revenue categories

---

### 📌 What this does

* `groupby("Region")` → groups transactions by region
* `sum()` → computes total revenue per region
* `idxmax()` → finds region name with highest revenue
* `idxmin()` → finds region name with lowest revenue

---

### 📊 Expected output format

---

## 📌 What this does

* Converts `Date` into proper datetime format
* Extracts month number (1–12)
* Groups data by month
* Computes total revenue per month

---
